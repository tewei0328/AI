# 1.取得鐵達尼號完整資料集

```
import pandas as pd
```

```
titanic=pd.read_csv('https://raw.githubusercontent.com/tewei0328/data/master/titanic.csv')
```

```
titanic.shape
```

```
titanic.head()
```

# 2.將資料切片成題目X, 和答案y
並進一步切割成訓練集, 和測試集

```
X=titanic[['pclass','age','sex']]
```

```
X.head()
```

```
y=titanic[['survived']]
```

```
y.head()
```

```
X.info()
```

# 3.補缺值

```
X['age'].fillna(X['age'].mean(),inplace=True)
```

```
X.info()
```

```
X.describe()
```

```
X.head()
```

# 4. 將資料進一步切割成訓練集, 和測試集

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=33)
```

```
X_train.head()
```

```
X_train.shape
```

```
X_test.shape
```

# 5. 取特徵

```
from sklearn.feature_extraction import DictVectorizer
```

```
vec=DictVectorizer(sparse=False)
```

```
X_train1=vec.fit_transform(X_train.to_dict(orient='records'))
```

```
X_train1
```

```
print(vec.feature_names_)
```

```
print(vec.get_feature_names())
```

```
X_test1=vec.transform(X_test.to_dict(orient='records'))
```

#6. 訓練模型

```
from sklearn.tree import DecisionTreeClassifier
```

```
dtc=DecisionTreeClassifier()
```

```
dtc.fit(X_train1,y_train)
```

#7. 預測結果

```
y_predict=dtc.predict(X_test1)
```

```
from sklearn.metrics import classification_report
```

```
print(dtc.score(X_test1, y_test))
```

```
print(classification_report(y_predict, y_test, target_names=['died','survived']))
```
