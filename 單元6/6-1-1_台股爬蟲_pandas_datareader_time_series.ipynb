```
import pandas_datareader as pdr
df_2330 = pdr.DataReader('2330.TW', 'yahoo')
df_2330.plot()
```

```
startTime = '2018-10-01'
endTime = '2018-10-30'
df_2330 = pdr.DataReader('2330.TW', 'yahoo', startTime, endTime)
df_2330.plot()
```
