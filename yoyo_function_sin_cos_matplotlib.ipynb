```python
import numpy as np 
import matplotlib.pyplot as plt 
X = np.linspace(0, 2 * np.pi, 100) 
YSinValues = np.sin(X) 
YCosValues = np.cos(X) 
plt.plot(X, YSinValues) 
plt.plot(X, YCosValues) 
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
X = np.linspace(0, 2 * np.pi, 100) 
Y1 = np.sin(X) 
Y2 = np.sin(X)*np.sin(X) 
Y3 = -np.cos(X)
Y = np.sin(X)*np.sin(X) -np.cos(X) 

plt.figure(figsize=(10, 7))
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi,np.pi*1.5,np.pi*2],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$',r'$+3\pi/2$',r'$+2\pi$'])

plt.plot(X, Y1)
plt.plot(X, Y2)
plt.plot(X, Y3)  
plt.plot(X, Y)  
plt.legend(['y=sinx','y=(sinx)^2','y=-cosx','y=(sinx)^2-cosx'])
plt.show()
```
