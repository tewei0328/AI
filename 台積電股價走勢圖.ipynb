# 台積電股價走勢圖畫法

本筆記本展示如何使用 `yfinance` 抓取台積電 (2330.TW) 的股價資料並繪製走勢圖。

```python
!pip install yfinance
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 設定股票代碼
ticker = "2330.TW"

# 下載最近一年的資料
df = yf.download(ticker, period="1y")

# 繪製走勢圖
plt.figure(figsize=(12, 6))
df['Close'].plot(title='TSMC (2330.TW) Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Price (TWD)')
plt.grid(True)
plt.show()
```
